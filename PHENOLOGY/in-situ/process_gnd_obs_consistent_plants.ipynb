```python
import os
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#before using code, retrieve the Plant Phenology Observations Realease-2025 from ...
#NEON (National Ecological Observatory Network). Plant phenology observations (DP1.10055.001), RELEASE-2025. https://doi.org/10.48443/6e2g-9e81. Dataset accessed from https://data.neonscience.org/data-products/DP1.10055.001/RELEASE-2025 on July 19, 2024.

#create a list of decid broadleaf individuals
#create a data frame to store individual plant data

#find all deciduous broadleaf individuals sampled in march-june of the year
#hard-coded filenames for processing TREE site on my file system; adjust folder and file names for your file system and site of interest
plants_list = os.path.join(os.path.expanduser("~"), 'Documents', 'work', 'NEON', 'ENSO', 'NEON_obs-phenology-plant 2','NEON.D05.TREE.DP1.10055.001.2015-07.basic.20250129T000730Z.RELEASE-2025','NEON.D05.TREE.DP1.10055.001.phe_perindividual.basic.20241118T174208Z.csv')

#print(trees_fns[0]) 
plantIDs = []
species = []
plot = []
some_plants = pd.read_csv(plants_list)
some_trees = some_plants[some_plants.growthForm == 'Deciduous broadleaf']
for i in range(len(some_trees)):
    the_tree = some_trees.iloc[i]
    plantIDs.append(the_tree.individualID)
    species.append(the_tree.scientificName)
    plot.append(the_tree.plotID)
print(len(plantIDs))
#print(some_trees.head)
d = {'plantID': plantIDs, 'species': species, 'plotID': plot, 'numyrs': ''}
the_trees = pd.DataFrame(data=d)

#populate list with years each individual is observed
#for each year
for year in range(2015,2025):
    data_dir = os.path.join(os.path.expanduser("~"), 'Documents', 'work', 'NEON', 'ENSO', 'NEON_obs-phenology-plant 2')
    filename_search = '*'+str(year)+'-0[3-6]*/*statusintensity*'
    obs_fns = sorted(glob(os.path.join(data_dir, filename_search)))
    #loop over the trees of interest
    for i in range(len(the_trees)):
        the_plantID = the_trees.iloc[i].plantID
        is_observed = 0
        #for each spring month
        for fn_ind in range(0,len(obs_fns)):
            #read in the obs for the month
            some_obs = pd.read_csv(obs_fns[fn_ind])
            the_obs = some_obs[(some_obs.individualID == the_plantID)]
            if len(the_obs) > 0:
                is_observed = 1
        if is_observed:
            the_trees.iat[i,3] = the_trees.iloc[i].numyrs + ' ' +str(year)

#look at it
print(the_trees.head)
the_trees.to_csv('TREE_pheno_individuals.csv')#, mode='a')
```

```python
# for individuals observed 2015-2024, find phenometrics
pheno_individuals_file = os.path.join(os.path.expanduser("~"), 'Documents', 'work', 'NEON', 'ENSO', 'TREE_pheno_individuals.csv')
some_plants = pd.read_csv(pheno_individuals_file)
#create data frames to store phenometrics for each individual across years
df = some_plants.copy()
for year in range(2015,2025):
    df[str(year)+' budbreak'] = np.nan
    df[str(year)+' full-leaf out'] = np.nan
    df[str(year)+' full canopy'] = np.nan
data_dir = os.path.join(os.path.expanduser("~"), 'Documents', 'work', 'NEON', 'ENSO', 'NEON_obs-phenology-plant 2')

for year in range(2015,2025):
    #find DOYs for earliest 3-10 "breaking leaf buds" and earliest one or more "Leaves" and earliest "increasing leaf size" >=95% of full size for each individual
    filename_search = 'NEON.D05.TREE.*'+str(year)+'-0[3-6]*/*statusintensity*'
    obs_fns = sorted(glob(os.path.join(data_dir, filename_search)))
    for fn_ind in range(0,len(obs_fns)):
        #read in the obs for the month
        some_obs = pd.read_csv(obs_fns[fn_ind])
        #loop over the trees of interest
        for i in range(len(df)):
            the_plantID = df.iloc[i].plantID
            #print(the_plantID)
            #find all observations of 3-10 "breaking leaf buds" for this tree; store earliest date in "budbreak" column if earlier than existing date
            the_obs = some_obs[(some_obs.individualID == the_plantID) & (some_obs.phenophaseName == 'Breaking leaf buds') & ( (some_obs.phenophaseIntensity == '3 to 10') | (some_obs.phenophaseIntensity == '11 to 100') | (some_obs.phenophaseIntensity == '101 to 1000') | (some_obs.phenophaseIntensity == '1001 to 10000'))]
            the_obs = the_obs.sort_values(by="dayOfYear")
            if len(the_obs) > 0:
                first_obs = the_obs.iloc[0].dayOfYear
                if ~(first_obs > df.loc[i,str(year)+' budbreak']):
                    df.at[i,str(year)+' budbreak'] = first_obs
            #find all observations of one or more "Leaves" for this tree; store earliest date in "full-leaf out" column if earlier than existing date
            the_obs = some_obs[(some_obs.individualID == the_plantID) & (some_obs.phenophaseName == 'Leaves') & (some_obs.phenophaseStatus == 'yes')]
            the_obs = the_obs.sort_values(by="dayOfYear")
            if len(the_obs) > 0:
                first_obs = the_obs.iloc[0].dayOfYear
                if ~(first_obs > df.loc[i,str(year)+' full-leaf out']):
                    df.at[i,str(year)+' full-leaf out'] = first_obs
            #find all observations of >=95% of full size "increasing leaf size" for this tree; store earliest date in "fullcanopy" column if earlier than existing date
            the_obs = some_obs[(some_obs.individualID == the_plantID) & (some_obs.phenophaseName == 'Increasing leaf size') & (some_obs.phenophaseIntensity == '>= 95%') ]
            the_obs = the_obs.sort_values(by="dayOfYear")
            if len(the_obs) > 0:
                first_obs = the_obs.iloc[0].dayOfYear
                if ~(first_obs > df.loc[i,str(year)+' full canopy']):
                    df.at[i,str(year)+' full canopy'] = first_obs

print(df)
df.to_csv('TREE_phenometrics.csv')
```
